```python
import pandas as pd
from datetime import datetime
import seaborn
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"..\data\price_data.csv", index_col=0)
df["timestamp"] = pd.to_datetime(df["timestamp"])
df.set_index("timestamp", inplace=True)
df.head()
```

```python
seaborn.heatmap(df.corr(), cmap = "Blues")
```

```python
plt.figure(figsize=(10, 5))
plt.xlabel("Date")
plt.ylabel("Price (JPY)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.plot(df[["Bitcoin_price"]])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.xlabel("Date")
plt.ylabel("Price (JPY)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.plot(df[["Financie Token_price"]])
plt.show()
```

# 

```python
log_rt_df = np.log(df).diff()
log_rt_df.head()
```

```python
seaborn.heatmap(log_rt_df.corr(), cmap = "Blues")
```

```python
mean_df = log_rt_df.mean()
std_df = log_rt_df.std()

mean_std_df = pd.concat([mean_df, std_df], axis=1)
mean_std_df.columns = ["mean", "std"]
mean_std_df
```

```python


fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.scatter(mean_std_df["std"], mean_std_df["mean"])
ax.grid()
ax.set_title('Mean vs Std')
ax.set_xlabel('std')
ax.set_ylabel('mean')

ax.plot()
```
